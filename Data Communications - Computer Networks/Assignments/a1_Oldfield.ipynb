```python
class Message:
    def __init__(self, size, num):
        self.size = size  # size of packets in bits
        self.num = num    # number of packets
        self.cord = 0
# Set messages and transmission rate
ms = [Message(4000, 4), Message(2000, 8), Message(1000, 16),
      Message(500, 32), Message(250, 64), Message(125, 128)]
#m = 16000   # message size
rate = 4000 # transmission rate

# Get transmission delay
def getDelay(m, ov):
    return (m.size + ov)/rate

# Simulate transmission
def getEnd(m, ov):
    return (m.num + 2)* getDelay(m, ov)

# Print results
def result(ms, ov):
    for x in ms:
        x.cord = getEnd(x, ov)
        print("\t", x.size, "bits takes", "%.2f" % getEnd(x, ov), "seconds.")

# Call
print("Results for 0 bits of overhead:")
result(ms, 0)
a = [x.cord for x in ms]

print("Results for 10 bits of overhead:")
result(ms, 10)
b = [x.cord for x in ms]

print("Results for 20 bits of overhead:")
result(ms, 20)
c = [x.cord for x in ms]
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
dom = [x.size for x in ms]

ax.tick_params(top="true", right="true")
ax.tick_params(axis="y",direction="in")
ax.tick_params(axis="x",direction="in")

ax.plot(dom, a, '-b+', label='Overhead= 0bits', linewidth=.5)
ax.plot(dom, b, '-r*', label='Overhead= 10bits', linewidth=.5)
ax.plot(dom, c, '-yx', label='Overhead= 20bits', linewidth=.5)

plt.gca().invert_xaxis()
ax.grid()
plt.legend()
```
